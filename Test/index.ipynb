```python
import pandas as pd
data = pd.read_csv('Dataset//zomato.csv')
```

```python

data.shape
```

```python
#split where there is + in phone no and then delte it
data['phone']=data['phone'].str.extract(r'\+91\s*(\d+)')
```

```python
# data.dropna(subset=['phone'], inplace=True)
data.dropna(subset=['phone'],inplace=True)
```

```python
data.shape
```

```python
#drop the data in which both of re no

data = data[~((data['book_table']=='No') & (data['online_order']=='No'))]
```

```python
data.head()
```

```python
data.groupby('online_order').size()
```

```python
# data["Client-Id"] = data.index

#reset index make client id in front
data.set_index('Client-Id', inplace=True)
```

```python
data.head()
```

```python
#drop column ok
data.drop(columns=['ok'],inplace=True)

#make new column simple



```

```python
#split the adress column into city and country

data[['zipcode', 'city', 'country', 'other']]=data['address'].str.split(',',expand=True,n=3)
```

```python
# data.drop('city', axis=1, inplace=True)
data.drop('zipcode', axis=1, inplace=True)
```

```python
data.head()
```

```python
#want to make country column name to address
data.rename(columns={'other':'address'}, inplace=True)
```

```python
#want to make adress 2nd column in dataset
cols = data.columns.tolist()
cols = cols[:1] + ['address'] + cols[1:-1]
data = data[cols]
```

```python
data.head()
```

```python
data.shape
```

```python
# Split on \n and create two columns
#data.dropna(subset=['reviews_list'], inplace=True)
 #Create columns one by one
data['rating_part0'] = data['reviews_list'].str.split('\n').str[0]
data['review_part1'] = data['reviews_list'].str.split('\n').str[1]  # Will be NaN if no second part
```

```python
data.drop(columns=['review_part1'], inplace=True)
```

```python
data.head(20)
```

```python

data = data.dropna(subset=['name', 'online_order', 'book_table', 'rate', 'votes', 'phone', 'location', 'rest_type', 'cuisines'])
```

```python
#export data
data.to_csv('Dataset//zomatocleaned.csv', index=True)
```
