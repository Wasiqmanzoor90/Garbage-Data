```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score

df = pd.read_csv("/home/wasiq/Documents/Project/My-Dataset/Data/Iris.csv")
df.head()
```

```python
le = LabelEncoder()
df['Species'] = le.fit_transform(df['Species'])
```

```python
X = df.iloc[:, 1:5]  # sepal and petal measurements
y = df['Species']    # encoded target column
```

```python
df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
model = LogisticRegression(max_iter=200)
model.fit(X_train, y_train)
```

```python
# Evaluate
y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
# Use X_test column names
new_flower = pd.DataFrame([[5.1, 3.5, 1.4, 0.2]], columns=X_test.columns)
prediction = model.predict(new_flower)
predicted_species = le.inverse_transform(prediction)
print(f"Predicted Species: {predicted_species[0]}")
```

```python
import pickle

with open('pickle_logistic.pkl', 'wb') as file:
    pickle.dump(model, file)

with open('label_encoder.pkl', 'wb') as file:
    pickle.dump(le, file)
```
