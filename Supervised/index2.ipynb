```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
#load data
data = pd.read_csv("Dataset/companies.csv")

```

```python
data.shape
```

```python
#data already cleaned
data.head()
```

```python
# transformation of string label into numerical label
lb = LabelEncoder()

data["Total_reviews"]=lb.fit_transform(data["Total_reviews"]) # convert object to numerical
data["Avg_salary"]=lb.fit_transform(data['Avg_salary'])# conver object to numerical
data["Interviews_taken"]=lb.fit_transform(data['Interviews_taken'])# conver object to numerical
data["Total_jobs_available"]=lb.fit_transform(data['Total_jobs_available'])# conver object to numerical
data["Total_benefits"]=lb.fit_transform(data['Total_benefits'])# conver object to numerical
```

```python
# spliting dependent and independent values
X = data.iloc[:, [ 5, 6, 7, 8, 9]]
Y = data["Ratings"]
```

```python

#train test split
X_train ,X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=0)
```

```python
#use logistic regrssion model


model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
# Just predictions
rating = pd.DataFrame(model.predict(X_test), columns=['Predicted_Rating'])
print(rating.head(20))
```

```python
rating2 = pd.DataFrame([[424, 720, 206, 296, 415]], columns=X_test.columns)
prediction = model.predict(rating2)

print(f"Predicted Rating: {prediction[0]:.2f}")
```
