```python
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv('Dataset//population.csv')
```

```python
top5 = data.sort_values('Rank', ascending=True).head(5)
```

```python
top5
```

```python
plt.bar(top5['Country/Territory'], top5['2022 Population'])
plt.show()
```

```python
#get particular one state
india = top5[top5['Country/Territory'] == 'India']
china = top5[top5['Country/Territory'] == 'China']
```

```python
india.iloc[0, 6:14].plot(kind='line', marker='o', label='India')
china.iloc[0, 6:14].plot(kind='line',marker='o', label='China')
years = [2022, 2020, 2015, 2010, 2000, 1990, 1980, 1970]
plt.xticks(range(8), years)  # Set clean year labels
plt.xlabel('Years')
plt.ylabel('Population')
plt.legend()  # Show country names
plt.show()
```

```python
india.iloc[0, 6:14].plot(kind='bar')
years = [2022, 2020, 2015, 2010, 2000, 1990, 1980, 1970]
plt.xticks(range(8), years)
plt.xlabel('Years')
plt.legend()  # Show country names
plt.show()
```

```python
top30 = data.sort_values('Rank',ascending=True).head(30)
```

```python
top30.set_index('Rank',inplace=True)
```

```python
plt.scatter(top30.index, top30['2020 Population'], color='darkred')
# Add labels and title
plt.xlabel('Rank')
plt.ylabel('Population')
plt.title('Population vs Rank (Top 30)')
plt.grid(True)
```

```python
top = data.sort_values('Rank', ascending=True).head(5)
```

```python
top
```

```python


plt.pie(
    top['2020 Population'],
    labels=top['Country/Territory'],
    autopct='%1.1f%%',
    explode=[0.05, 0.05, 0.05, 0.05, 0.05]  # slightly separate slices
)
plt.title('Population Share of Top 5 Countries (2020)')
plt.show
```
