```python
import pandas as pd
data = pd.read_csv('Dataset/Mobile.csv')
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
data['brand']= lb.fit_transform(data['brand'])
data['model']= lb.fit_transform(data['model'])
data['country']= lb.fit_transform(data['country'])
data['verified_purchase']= lb.fit_transform(data['verified_purchase'])
```

```python
data.head()
```

```python
X=data.iloc[:, [2, 3, 4, 5, 8, 12, 15, 16, 17, 18, 19, 20, 22, 23]]
Y=data["rating"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_tain, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, Y_tain)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

y_pred = model.predict(X_test)

print("R² Score:", r2_score(Y_test, y_pred))
print("Mean Squared Error:", mean_squared_error(Y_test, y_pred))
```

```python
# Make sure result has the SAME columns as X
result = pd.DataFrame([[18, 4, 15, 307.8, 5.70,  4, 1, 3, 2, 4, 3, 2, 12, 5]], 
                      columns=X.columns)  # This must match X.columns EXACTLY

prediction = model.predict(result)
print(f"Rating: {prediction[0]:.2f}")
```
