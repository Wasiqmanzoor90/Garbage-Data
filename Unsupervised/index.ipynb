```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('/home/wasiq/Documents/Project/My-Dataset/Data/Iris.csv')
```

```python
data.head()
```

```python
plt.scatter(data['SepalLengthCm'], data['SepalWidthCm'])
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Sepal Length vs Width')
plt.show()
```

```python
from sklearn.cluster import KMeans

sse=[]
k_range = range(1,11)
for k in k_range:
    km=KMeans(n_clusters=k)
    km.fit(data[['SepalLengthCm','SepalWidthCm']])
    # plt.scatter(data['PetalLengthCm'],data['PetalWidthCm']
    sse.append(km.inertia_)
plt.plot(k_range,sse, marker="o")
```

```python
#agglomerative
from sklearn.cluster import AgglomerativeClustering
agg = AgglomerativeClustering(n_clusters=3)
data['Agg_Labels'] = agg.fit_predict(data[['SepalLengthCm','SepalWidthCm']])
plt.scatter(data['SepalLengthCm'], data['SepalWidthCm'], c=data['Agg_Labels'])
# plt.xlabel('Sepal Length (cm)')
# plt.ylabel('Sepal Width (cm)')
# plt.title('Agglomerative Clustering')
# plt.show()
```
