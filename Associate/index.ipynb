```python
import pandas as pd
df = pd.read_csv('/home/wasiq/Documents/Project/My-Dataset/Data/Market_Basket.csv')
```

```python
df.shape
```

```python
#convert null values to 0
df.fillna(0,inplace=True)
```

```python
#only ist 100` rows`
data = df.head(2000)
```

```python
transaction = []
for i in range(0,len(data)):
    transaction.append([str(data.values[i,j]) for j in range(0,20) if str(data.values[i,j])!='0' ])
```

```python
transaction[4]
```

```python
#!pip install apyori
from apyori import apriori
rules=apriori(transactions=transaction,min_support=0.003,min_confidence=0.4,min_lift=3,min_length=2,max_length=4)
```

```python
Result=list(rules)
len(Result)
```

```python
Result[0]
```

```python
for items in Result:
    pair = items[0]
    for itemss in pair:
        print(itemss)
        
    support = items[1]
    confidence = items[2][0][2]
    lift=items[2][0][3]

    print("Support = ",support)
    print("Confidence = ",confidence)
    print("Lift = ",lift)
    print("----------------------------------------------------------------------------------------------------------------")
```
